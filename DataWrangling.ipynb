### Transforming the canada dataframe, adding a new column that shows percentage increase or decrease

```python
level = np.unique(canada.Levelofstudy)
stats = np.unique(canada.Statistics)
```

```python
def get_measures(level, stats, df):
    level_dict = {}
    stats_dict = {}
    for ele in level:
        for sta in stats:
            stats_dict[sta] = list(df[(df.Statistics == sta) & (df.Levelofstudy == ele)].Value)
        level_dict[ele] = dict(stats_dict)
    return level_dict
```

```python
measures = get_measures(level,stats, canada)
```

```python
print(measures)
```

```python
def percent_inc_dec(y,x):
    return ((y-x)/x)*100 if x != 0 else 0
```

```python
def cal_percent(level,stats,measures):
    level_dict = {}
    stats_dict = {}
    for ele in level:
        for sta in stats:
            tmp_list = [0]
            for i in range(len(measures[ele][sta])-1):
                tmp_list.append(percent_inc_dec(measures[ele][sta][i+1],measures[ele][sta][i]))
            stats_dict[sta] = tmp_list
        level_dict[ele] = dict(stats_dict)
    
    return level_dict
```

```python
percentages = cal_percent(level,stats,measures)
```

```python
print(percentages)
```

```python
canada['Percent'] = pd.Series(dtype='float64')
canada
```

```python
pd.set_option('mode.chained_assignment', None)
```

```python
def get_percents(level,stats,percentages, df):
    for ele in level:
        for sta in stats:
            index_list = list(df[(df.Statistics == sta) & (df.Levelofstudy == ele)].index)
            for val_index,val in enumerate(index_list):
                value = percentages[ele][sta][val_index]
                df.loc[val,'Percent'] = value
    return df
```

```python
canada = get_percents(level,stats,percentages, canada)
```

```python
val = canada[canada.Statistics == 'Number of graduates'].Percent
year = canada[canada.Statistics == 'Number of graduates'].Year
level = canada[canada.Statistics == 'Number of graduates'].Levelofstudy
```

```python
sb.lineplot(x = year, y = val, hue = level)
plt.show()
```
